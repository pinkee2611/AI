```python
import matplotlib.pyplot as plt
from keras.models import load_model
from matplotlib.image import imread
folder = '/content/drive/MyDrive/tien/'
for i in range(8):
  plt.subplot(10,6,i+1)
  filename=folder+'100k ('+str(i)+').jpg'
  img=imread(filename)
  plt.imshow(img)
plt.show()
```

```python
from os import listdir
from numpy import asarray,save
from keras.utils import load_img, img_to_array
folder = '/content/drive/MyDrive/tien/'
photos,labels=list(),list()
for file in listdir(folder):
  output=0.0
  if file.startswith('5k'):
    output=1.0
  if file.startswith('10k'):
    output=2.0
  if file.startswith('20k'):
    output=3.0
  if file.startswith('50k'):
    output=4.0
  if file.startswith('100k'):
    output=5.0  
  if file.startswith('500k'):
    output=6.0  
  img=load_img(folder+file, target_size=(30,40))
  photo=img_to_array(img)
  photos.append(photo)
  labels.append(output)
photos=asarray(photos)
labels=asarray(labels)
print(photos.shape,labels.shape)
save('gia_tri_tien_photos.npy',photos)
save('gia_tri_tien_labels.npy',labels)
```

```python
import numpy as np
x_train = np.load('/content/gia_tri_tien_photos.npy')
y_train = np.load('/content/gia_tri_tien_labels.npy')
print(x_train.shape)
print(y_train.shape)
x=x_train
y=y_train
from keras.utils import to_categorical
x_train = x_train.astype('float32')/255
y_train = to_categorical(y_train,100)
```

```python
from keras.backend import flatten
from keras.models import Sequential
from keras.layers import Dense,Dropout,Flatten,Conv2D,MaxPooling2D,Normalization,LeakyReLU
from keras.optimizers import Adam
#32 lần
model = Sequential()
model.add(Conv2D(32,kernel_size = (3,3),activation = 'relu',input_shape=(30,40,3),padding='Same'))
model.add(MaxPooling2D((2,2),padding='same'))
model.add(Dropout(0.25))

#64 lần
model.add(Conv2D(64,(3,3),activation ='relu',padding ='same'))
model.add(MaxPooling2D((2,2),padding='same'))
model.add(Dropout(0.25))

#128 lần
model.add(Conv2D(128,(3,3),activation ='relu',padding ='same'))
model.add(MaxPooling2D((2,2),padding='same'))
model.add(Dropout(0.25))

#256 lần
model.add(Conv2D(256,(3,3),activation ='relu',padding ='same'))
model.add(MaxPooling2D((2,2),padding='same'))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(256,activation='relu'))
model.add(LeakyReLU(alpha=0.1))
model.add(Dense(100,activation='softmax'))

from keras.losses import categorical_crossentropy
model.compile(loss='categorical_crossentropy',optimizer=Adam(),metrics=['accuracy'])
model.summary()
train = model.fit(x_train,y_train,epochs=200,batch_size=64,verbose=1)
```

```python
model.save('gia_tri_tien.h5')
```
